*LOADING DATASET*

```python
import pandas as pd

df = pd.read_csv(r"C:\Users\janan\Downloads\archive\telco_cleaned.csv") 

# Clean missing values if any
df.dropna(inplace=True)
df.head()
df.columns
```

*DROP IRRELEVENT COLUMNS*

```python
df = df.drop(columns=[
    'Customer ID', 'Customer Status', 'Churn Label', 'Churn Score', 
    'Churn Category', 'Churn Reason'
])
```

*ENCODE CATEGORICAL VARIABLES*

```python
from sklearn.preprocessing import LabelEncoder

label_encoders = {}
for col in df.select_dtypes(include='object').columns:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col].astype(str))
    label_encoders[col] = le
```

*TRAIN A MODEL TO PREDICT CHURN*

```python
df_full = pd.read_csv(r"C:\Users\janan\Downloads\archive\telco_cleaned.csv")
target = LabelEncoder().fit_transform(df_full['Churn Label'])  # Yes = 1, No = 0
features = df.dropna()
```

*SPLIT AND TRAIN*

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2)

model = RandomForestClassifier(n_estimators=100)
model.fit(X_train, y_train)
```

*GENERATING PREDICTIONS*

```python
df_full = df_full.dropna().copy()
df_full['Churn_Probability'] = model.predict_proba(features)[:, 1]

df_full['Churn_Risk'] = pd.cut(df_full['Churn_Probability'], 
                               bins=[0, 0.4, 0.7, 1], 
                               labels=['Low', 'Medium', 'High'])
```
